```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cross_validation import train_test_split
from sklearn import naive_bayes
from sklearn.metrics import roc_auc_score
```

```python
df=pd.read_csv('twitter_sentiment.csv',encoding = "ISO-8859-1")
df.head()
```

```python
vectorizer=TfidfVectorizer(use_idf=True, lowercase=True, strip_accents='ascii',stop_words='english')
y=df.Sentiment
x=vectorizer.fit_transform(df.SentimentText)
print(y.shape)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,random_state=25)
clf=naive_bayes.MultinomialNB()
clf.fit(X_train,Y_train)
roc_auc_score(Y_test,clf.predict_proba(X_test)[:,1])
```

```python
movie_reviews_array=np.array(["Thugs of Hindustan is just another worst"])
movie_review_vector=vectorizer.transform(movie_reviews_array)
op=clf.predict(movie_review_vector)
if op==0:
    print("This is a negative(-ve) comment")
else:
    print("This is a positive(+ve) comment")
```
